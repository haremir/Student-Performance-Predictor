```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
data_path = os.path.join('..', 'data', 'raw', 'StudentsPerformance.csv')
df = pd.read_csv(data_path)
```

```python
print("Veri boyutu:", df.shape)
print("\nİlk 5 satır:")
print(df.head())
```

```python
print("\nVeri tiplerine bakalım:")
print(df.dtypes)
```

```python
print("\nEksik değer kontrolü:")
print(df.isnull().sum())
```

```python
print("\nBetimsel istatistikler:")
print(df.describe())
```

```python
# Kategorik değişkenlerin dağılımı
plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
sns.countplot(x='gender', data=df)
plt.title('Cinsiyet Dağılımı')

plt.subplot(2, 3, 2)
sns.countplot(x='race/ethnicity', data=df)
plt.title('Etnik Köken Dağılımı')
plt.xticks(rotation=45)

plt.subplot(2, 3, 3)
sns.countplot(x='parental level of education', data=df)
plt.title('Ebeveyn Eğitim Düzeyi')
plt.xticks(rotation=45)

plt.subplot(2, 3, 4)
sns.countplot(x='lunch', data=df)
plt.title('Öğle Yemeği Tipi')

plt.subplot(2, 3, 5)
sns.countplot(x='test preparation course', data=df)
plt.title('Test Hazırlık Kursu')

plt.tight_layout()
plt.savefig(os.path.join('..', 'notebooks', 'images', 'categorical_distributions.png'))
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.histplot(df['math score'], kde=True)
plt.title('Matematik Puanları Dağılımı')

plt.subplot(1, 3, 2)
sns.histplot(df['reading score'], kde=True)
plt.title('Okuma Puanları Dağılımı')

plt.subplot(1, 3, 3)
sns.histplot(df['writing score'], kde=True)
plt.title('Yazma Puanları Dağılımı')

plt.tight_layout()
plt.savefig(os.path.join('..', 'notebooks', 'images', 'score_distributions.png'))
```

```python
plt.figure(figsize=(10, 8))
correlation = df[['math score', 'reading score', 'writing score']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Sınav Puanları Arasındaki Korelasyon')
plt.savefig(os.path.join('..', 'notebooks', 'images', 'score_correlation.png'))
```

```python
# Kategorik değişkenlerin puanlara etkisi
plt.figure(figsize=(15, 20))

# Cinsiyet ve puanlar
plt.subplot(5, 1, 1)
sns.boxplot(x='gender', y='math score', data=df)
plt.title('Cinsiyet ve Matematik Puanları')

plt.subplot(5, 1, 2)
sns.boxplot(x='race/ethnicity', y='math score', data=df)
plt.title('Etnik Köken ve Matematik Puanları')

plt.subplot(5, 1, 3)
sns.boxplot(x='parental level of education', y='math score', data=df)
plt.title('Ebeveyn Eğitim Düzeyi ve Matematik Puanları')
plt.xticks(rotation=45)

plt.subplot(5, 1, 4)
sns.boxplot(x='lunch', y='math score', data=df)
plt.title('Öğle Yemeği Tipi ve Matematik Puanları')

plt.subplot(5, 1, 5)
sns.boxplot(x='test preparation course', y='math score', data=df)
plt.title('Test Hazırlık Kursu ve Matematik Puanları')

plt.tight_layout()
plt.savefig(os.path.join('..', 'notebooks', 'images', 'categorical_vs_math.png'))
```

```python
df['total score'] = df['math score'] + df['reading score'] + df['writing score']
df['average score'] = df['total score'] / 3

plt.figure(figsize=(10, 6))
sns.histplot(df['average score'], kde=True)
plt.title('Ortalama Puan Dağılımı')
plt.savefig(os.path.join('..', 'notebooks', 'images', 'average_score.png'))
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='parental level of education', y='average score', hue='test preparation course', data=df)
plt.title('Ebeveyn Eğitim Düzeyi, Test Hazırlık Kursu ve Ortalama Puan')
plt.xticks(rotation=45)
plt.savefig(os.path.join('..', 'notebooks', 'images', 'education_prep_score.png'))
```
